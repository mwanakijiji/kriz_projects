```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from astropy.io import fits

%matplotlib qt
```

```python
stem = "/Users/bandari/Documents/git.repos/kriz_projects/notebooks_for_development/output_other/translate_apertures_test_210826/"
fits_stem = "/Users/bandari/Documents/git.repos/kriz_projects/data/2021-08-04/raw/"
```

```python
# names of .dat files

file_names_photometry = sorted(glob.glob(stem + "aper_photom_V1432_Aql-*dat"))
```

```python
frames = []
names = []
frame_columns = ['id', 'xcenter', 'ycenter', 'aperture_sum']

datacube_photometry = np.nan*np.ones((len(file_names_photometry),302,6))
#print(datacube_photometry.shape)

list_fit_file_names = []
list_jd = []
list_helio_jd = []

for num_frame in range(0,len(file_names_photometry)):
    
    # read in DataFrame
    new_frame = pd.read_csv(file_names_photometry[num_frame], delim_whitespace=True)
    new_frame_np = new_frame.to_numpy()
    print(file_names_photometry[num_frame])
    
    datacube_photometry[num_frame,:,:] = new_frame_np
    
    # also read in FITS file to extract the UT time, which we will add in a column
    fit_name = os.path.basename(file_names_photometry[num_frame]).split("photom_")[1].split(".")[0] + ".fit"
    hdul = fits.open(fits_stem + fit_name)
    list_fit_file_names.append(fit_name)
    list_jd.append(hdul[0].header['JD'])
    list_helio_jd.append(hdul[0].header['JD-HELIO'])
```

```python
# arrange into DataFrame

# numerical data
array_recast = np.nan*np.ones((len(datacube_photometry[:,0,0]),len(datacube_photometry[0,:,0])))

for num_star in range(0,len(datacube_photometry[0,:,0])):
    array_recast[:,num_star] = datacube_photometry[:,num_star,5]
    plt.plot(datacube_photometry[:,num_star,5])
plt.xlabel("Frame number (~5 hr observation duration)")
plt.ylabel("Direct counts (no sky subtraction)")
plt.show()
```

```python
# add numbers for stars

star_strings = map(str, np.arange(0,302)) 
```

```python
# convert to DataFrame

photometry_df = pd.DataFrame(array_recast, columns = star_strings)
```

```python
# put in file names of raw data, and JD epochs

photometry_df["fit_file_name"] = list_fit_file_names
photometry_df["jd-2459431"] = np.subtract(list_jd,2459431)
photometry_df["jd_helio-2459431"] = np.subtract(list_helio_jd,2459431)
```

```python
# write to csv

photometry_df.to_csv("test_time_series_20210917.csv")
```
